Импорт бибилиотек

```python
import time
import bs4  # импортируем Beautiful Soup
import re # регулярка для поиска цифр
import pandas as pd
from google.colab import files
```

### Очистка данных

```python
def read_url(url):
    """ Читает содержимое сайта по URL"""
    import urllib.request
    with urllib.request.urlopen(url) as webpage:
        text = webpage.read().decode('utf-8')
    return text
```

Открываем текстовый фаил со ссылками на все дома и приводм их к списку

```python
domatxt = open('/content/razumnoe-prostranstvo.txt', 'r')
listHouse = domatxt.read().split(sep='\n')
domatxt.close()
print(listHouse)
```

Проходим по списку загружая страницы и вытаскивая нужную информацию

```python
list_arr_b = []
list_arr_o = []
list_price = []
print(len(list_price))
print(len(list_arr_b))
print(len(list_arr_o))

print("Количество домов =", len(listHouse))
print("--------------------------------------")

counter = 0

for house in listHouse:
  text = read_url(house) #текст загруженной страницы
  bs = bs4.BeautifulSoup(text, 'html.parser') #создаем парсер

  # Площадь дома и площадь террасы (если имеется)
  area = bs.find_all("div", {"class": "project__aside__area"})
  # Цена
  price = bs.find_all("span", {"class": "current"})

  i_basic = str(area).find("Общая площадь:")
  i_other = str(area).find("Площадь террасы:")

  arr_b = arr_o = 0

  print("== " + str(counter) + " ==== " + str(house))

  str_arr_b = re.findall(r'\d+', str(area)[i_basic + 27 : i_basic + 27 + 8].replace(' ',''))[0]
  arr_b = int(str_arr_b)
  print("Общая: " + str_arr_b)

  if (i_other > 0) :
    str_arr_o = re.findall(r'\d+', str(area)[i_other + 29 : i_other + 29 + 8].replace(' ',''))[0]
    arr_o = int(str_arr_o)
    print("Терасса: " + str_arr_o)

  prc = int(str(price)[23: ][ :-12].replace(' ',''))
  print("Цена: " + str(price)[23: ][ :-12].replace(' ',''))


  arr = arr_b - arr_o # полезная площадь (вся за исключением терассы)

  unit = 0
  if (arr != 0) :
    unit = prc/arr

  print("Цена за м2 = " + str(unit))
  print("")

  list_arr_b.append(arr_b)
  list_arr_o.append(arr_o)
  list_price.append(prc)

  counter += 1
  time.sleep(1) #задержка для уменьшения нагрузки

print("|||||||||||||||||||||||||||||||||||||||||||||||||||||||||||")
print('Размеры полученных массивов')
print(len(list_price))
print(len(list_arr_b))
print(len(list_arr_o))

```

Сохраняем в таблицу эксель (в дальнейшем можно загрузить эту таблицу что бы не проводить полное сканирвоание повторно)

```python
df = pd.DataFrame({
    'link' : listHouse,
    'price': list_price,
    'arr_b': list_arr_b,
    'arr_o': list_arr_o
})
df.to_excel('../r_p.xlsx')
```
